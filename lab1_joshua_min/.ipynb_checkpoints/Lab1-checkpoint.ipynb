```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import neighbors, metrics
import seaborn as sns
from scipy.io import arff
%config IPCompleter.greedy=True
```

**PART A)**

```python
DATA_PATH = './column_2C.dat'
var_set = [
    "pelvic_incidence",
    "pelvic_tilt",
    "lumbar_lordosis_angle",
    "sacral_slope",
    "pelvic_radius",
    "grade_of_spondylolisthesis"
]
head_set = []
head_set.extend(var_set)
head_set.append('label')
df = pd.read_table(DATA_PATH, sep=' ', header=None, names=head_set)
df.columns = head_set
```

**PART B) i)**

```python
# seaborn pairplots to display
sns.pairplot(df, hue='label')
```

**PART B ii)**

```python
for var in var_set:
    sns.catplot(data=df, x='label', y=var, kind='box')
```

**PART B) iii) && PART C) i)**

```python
# Get first 70 rows of class 0 and first 140 rows of class 1
df.replace(['AB', 'NO'], [1, 0], inplace=True)
tr_1 = df[df['label']==1][:140]
t_1 = df[df['label']==1][140:]
tr_0 = df[df['label']==0][:70]
t_0 = df[df['label']==0][70:]
```

```python
X_1 = np.array(tr_1.drop(['label'], 1))
X_0 = np.array(tr_0.drop(['label'], 1))
y_1 = np.array(tr_1['label'])
y_0 = np.array(tr_0['label'])

X_train = np.concatenate([X_1, X_0])
y_train = np.concatenate([y_1, y_0])

X_1 = np.array(t_1.drop(['label'], 1))
X_0 = np.array(t_0.drop(['label'], 1))
y_1 = np.array(t_1['label'])
y_0 = np.array(t_0['label'])

X_test = np.concatenate([X_1, X_0])
y_test = np.concatenate([y_1, y_0])

X_1 = np.array(tr_1.drop(['label'], 1))
X_0 = np.array(tr_0.drop(['label'], 1))
y_1 = np.array(tr_1['label'])
y_0 = np.array(tr_0['label'])
```

**PART C) ii)**
- Confusion matrix, TPR, TNR, Precision, and F-Score when k=k*

```python
# TODO: BREAK TIES WITH F-SCORE
opt_k = None
confusion_matrix = None
tpr = None
tnr = None
results = None
perf1 = []

for k in range(208, 0, -3):
    clf = neighbors.KNeighborsClassifier(
        n_neighbors=k
    )
    clf.fit(X_train, y_train)
    results = clf.predict(X_test)
    
    # Find training error
    err = metrics.mean_squared_error(
        y_test, 
        results
    )
    
    # Get confusion matrix and tpr/tnr
    confusion_matrix = metrics.confusion_matrix(y_test, results)
    FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)  
    FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
    TP = np.diag(confusion_matrix)
    TN = confusion_matrix.sum() - (FP + FN + TP)
    tpr = TP/(TP+FN)
    tnr = TN/(TN+FP)
    
    # Record metrics for this k
    perf.append({
        "test_err": err,
        "k": k,
        "mtx": confusion_matrix,
        "f": metrics.f1_score(y_test, results),
        "recall": metrics.recall_score(y_test, results),
        "tpr": tpr,
        "tnr": tnr,
        "results": results,
        "train_err": clf.score(X_train, y_train)
    })
    
# Helper function for selecting best k based on:
# 1. Minimum test error
# 2. F1 score
# 3. Recall score
def best_k_metrics(x):
    return (x["test_err"], x["f"], x["recall"])

opt_k = min(perf1, key=best_k_metrics)
        
print("k*:", opt_k["k"])
print()
print("Confusion Matrix:")
print(opt_k["mtx"])
print()
print("True Positive Rate:")
print(f'\tClass 0: {opt_k["tpr"][0]}')
print(f'\tClass 1: {opt_k["tpr"][1]}')
print()
print("True Negative Rate:")
print(f'\tClass 0: {opt_k["tnr"][0]}')
print(f'\tClass 1: {opt_k["tnr"][1]}')
print()
print("Training Error:", opt_k["train_err"])
print()
print(metrics.classification_report(y_test, opt_k["results"]))
```

**PART C) iii)**
- Learning Curve Plot

```python
learning_curve = []
for N in range(10, 220, 10):
    # Partition test dataset
    Ny_train = np.concatenate([
        y_1[:N - N//3],
        y_0[:N//3]
    ])
    NX_train = np.concatenate([
        X_1[:N - N//3],
        X_0[:N//3]
    ])
    
    perf = []
    for k in range(1, N, 5):
        clf = neighbors.KNeighborsClassifier(
        n_neighbors=k
        )
        clf.fit(NX_train, Ny_train)
        results = clf.predict(X_test)
        
        # Find training error
        err = metrics.mean_squared_error(
            y_test, 
            results
        )

        # Get confusion matrix and tpr/tnr
        confusion_matrix = metrics.confusion_matrix(y_test, results)
        FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)  
        FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
        TP = np.diag(confusion_matrix)
        TN = confusion_matrix.sum() - (FP + FN + TP)
        tpr = TP/(TP+FN)
        tnr = TN/(TN+FP)

        # Record metrics for this k
        perf.append({
            "test_err": err,
            "k": k,
            "mtx": confusion_matrix,
            "f": metrics.f1_score(y_test, results),
            "recall": metrics.recall_score(y_test, results),
            "tpr": tpr,
            "tnr": tnr,
            "results": results,
            "train_err": clf.score(X_train, y_train)
        })
    best_k = min(perf, key=best_k_metrics)
    learning_curve.append([N, best_k["k"], best_k["test_err"]])
learning_curve = pd.DataFrame(
    learning_curve, 
    columns=["N", "k*","Test Error"]
)
#learning_curve.set_index('N', inplace=True)
sns.lineplot(learning_curve["N"], learning_curve["Test Error"])
```

**PART D)**
- Minkowski, Manhattan, Chebyshev, and Mahlanobis

```python
table = []


# MINKOWSKI
perf_minkowski = []
for k in range(1, 197, 1):
    clf = neighbors.KNeighborsClassifier(
        n_neighbors=k,
        metric='minkowski',
    )
    clf.fit(X_train, y_train)
    results = clf.predict(X_test)
    # Find training error
    err = metrics.mean_squared_error(
        y_test, 
        results
    )

    # Get confusion matrix and tpr/tnr
    confusion_matrix = metrics.confusion_matrix(y_test, results)
    FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)  
    FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
    TP = np.diag(confusion_matrix)
    TN = confusion_matrix.sum() - (FP + FN + TP)
    tpr = TP/(TP+FN)
    tnr = TN/(TN+FP)

    # Record metrics for this k
    perf_minkowski.append({
        "test_err": err,
        "k": k,
        "mtx": confusion_matrix,
        "f": metrics.f1_score(y_test, results),
        "recall": metrics.recall_score(y_test, results),
        "tpr": tpr,
        "tnr": tnr,
        "results": results,
        "train_err": clf.score(X_train, y_train)
    })
minkowski_k = min(perf_minkowski, key=best_k_metrics)
table.append(["Minkowski", "NA", minkowski_k["k"], minkowski_k["test_err"], minkowski_k["train_err"]])


#MANHATTAN
perf_manhattan = []
for k in range(1, 197, 1):
    clf = neighbors.KNeighborsClassifier(
        n_neighbors=k,
        metric='minkowski',
        p=1
    )
    clf.fit(X_train, y_train)
    # Find training error
    err = metrics.mean_squared_error(
        y_test, 
        results
    )

    # Get confusion matrix and tpr/tnr
    confusion_matrix = metrics.confusion_matrix(y_test, results)
    FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)  
    FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
    TP = np.diag(confusion_matrix)
    TN = confusion_matrix.sum() - (FP + FN + TP)
    tpr = TP/(TP+FN)
    tnr = TN/(TN+FP)

    # Record metrics for this k
    perf_manhattan.append({
        "test_err": err,
        "k": k,
        "mtx": confusion_matrix,
        "f": metrics.f1_score(y_test, results),
        "recall": metrics.recall_score(y_test, results),
        "tpr": tpr,
        "tnr": tnr,
        "results": results,
        "train_err": clf.score(X_train, y_train)
    })
manhattan_k = min(perf_manhattan, key=best_k_metrics)
table.append(["Manhattan", 1, manhattan_k["k"], manhattan_k["test_err"], manhattan_k["train_err"]])


# MINKOWSKI WITH LOG_10(p) in {0.1, 0.2... 1.0}
for i in np.arange(0.1, 1.1, 0.1):
    clf = neighbors.KNeighborsClassifier(
        n_neighbors=minkowski_k["k"],
        metric='minkowski',
        p=10**i
    )
    clf.fit(X_train, y_train)
    results = clf.predict(X_test)
    err = metrics.mean_squared_error(
        y_test, 
        results
    )
    table.append(["Minkowski", round(i, 1), minkowski_k["k"], err, clf.score(X_train, y_train)])

    
# CHEBYSHEV
clf = neighbors.KNeighborsClassifier(
    n_neighbors=minkowski_k["k"],
    metric='minkowski',
    p=float("inf")
)
clf.fit(X_train, y_train)
results = clf.predict(X_test)
err = metrics.mean_squared_error(
    y_test, 
    results
)
table.append(["Chebysheev", "INF", minkowski_k["k"], err, clf.score(X_train, y_train)])


# MAHALANOBIS
perf_maha = []
for k in range(1, 197, 1):
    clf = neighbors.KNeighborsClassifier(
        n_neighbors=k,
        algorithm='brute',
        metric='mahalanobis',
        metric_params={'V': np.cov(X_train)}
    )
    clf.fit(X_train, y_train)
    results = clf.predict(X_test)
    # Find training error
    err = metrics.mean_squared_error(
        y_test, 
        results
    )

    # Get confusion matrix and tpr/tnr
    confusion_matrix = metrics.confusion_matrix(y_test, results)
    FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)  
    FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
    TP = np.diag(confusion_matrix)
    TN = confusion_matrix.sum() - (FP + FN + TP)
    tpr = TP/(TP+FN)
    tnr = TN/(TN+FP)

    # Record metrics for this k
    perf_maha.append({
        "test_err": err,
        "k": k,
        "mtx": confusion_matrix,
        "f": metrics.f1_score(y_test, results),
        "recall": metrics.recall_score(y_test, results),
        "tpr": tpr,
        "tnr": tnr,
        "results": results,
        "train_err": clf.score(X_train, y_train)
    })
maha_k = min(perf_maha, key=best_k_metrics)
table.append(["Minkowski", "NA", maha_k["k"], maha_k["test_err"], maha_k["train_err"]])

d_results = pd.DataFrame(
    table, 
    columns=["Name", "P", "k*", "Test Error", "Train Score (Inverse of Training Error)"]
)
d_results
```

**PART E)**
- Weighted polling for Minkowski, Manhattan, and Chebyshev

```python
table = []

min_test_err = 1
opt_k = None
for k in range(1, 197, 1):
    clf = neighbors.KNeighborsClassifier(
        n_neighbors=k,
        weights='distance',
        metric='minkowski',
        p=1
    )
    clf.fit(X_train, y_train)
    results = clf.predict(X_test)
    err = metrics.mean_squared_error(
        y_test, 
        results
    )
    if err < min_test_err:
        opt_k = k
        min_test_err = err
table.append(["Manhattan", 1, opt_k, min_test_err])

min_test_err = 1
opt_k = None
for k in range(1, 197, 1):
    clf = neighbors.KNeighborsClassifier(
        n_neighbors=k,
        weights='distance',
        metric='minkowski',
        p=float("inf")
    )
    clf.fit(X_train, y_train)
    results = clf.predict(X_test)
    err = metrics.mean_squared_error(
        y_test, 
        results
    )
    if err < min_test_err:
        opt_k = k
        min_test_err = err
table.append(["Chebyshev", "INF", opt_k, min_test_err])

min_test_err = 1
opt_k = None
for k in range(1, 197, 1):
    clf = neighbors.KNeighborsClassifier(
        n_neighbors=k,
        weights='distance',
        metric='minkowski',
    )
    clf.fit(X_train, y_train)
    results = clf.predict(X_test)
    err = metrics.mean_squared_error(
        y_test, 
        results
    )
    if err < min_test_err:
        opt_k = k
        min_test_err = err
table.append(["Minkowski", "NA", opt_k, min_test_err])

d_results = pd.DataFrame(
    table, 
    columns=["Name", "P", "k*", "Test Error"]
)
d_results
```
