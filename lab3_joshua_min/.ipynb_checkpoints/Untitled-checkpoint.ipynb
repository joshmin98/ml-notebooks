```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import chi2
import sklearn.metrics as metrics
from sklearn.preprocessing import StandardScaler
from IPython.display import display, Markdown
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
```

## Part 1

### a/b) Selecting data

```python
label = 'diagnosis'
cols = [
    'id',
    label
]
for i in range(1, 31, 1):
    cols.append(f'feature_{str(i)}')
df = pd.read_csv('./wdbc.data', names=cols)
cols = cols[1:]
df.drop(columns=['id'], inplace=True)
df.diagnosis = df.diagnosis.map({'M': 1, 'B': 0})
feature_cols = [col for col in cols if col is not 'diagnosis']

malignant = df.loc[df['diagnosis'] == 1]
benign = df.loc[df['diagnosis'] == 0]

test_set = malignant[:50].append(benign[:50])
train_set = malignant[50:].append(benign[50:])
train_X = train_set.copy().drop(columns=[label])
train_y = train_set.copy().drop(columns=feature_cols)
test_X = test_set.copy().drop(columns=[label])
test_y = test_set.copy().drop(columns=feature_cols)
```

### c) Binary Classification for Logistic Regression

#### i) Scatterplots for features in the training set

```python
color_function = {0: "blue", 1: "red"}
colors = df[label].map(lambda x: color_function.get(x))
pd.plotting.scatter_matrix(df[feature_cols], c=colors, alpha = 0.5, figsize = (30, 30)); 
```

#### ii) Logistic Regression, and the reporting of confusion matrix, ROC, precision, recall, F1 score, and AUC for train and test sets.

```python
logr = LogisticRegression(random_state=1, solver='liblinear')
logr.fit(train_X, train_y[label])
pred = logr.predict(test_X)
pred_prob = logr.predict_proba(test_X)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(test_y, pred_prob)
true = test_y

plt.plot([0,1],[0,1], 'k--')
plt.plot(fpr, tpr, label='Logistic Regression')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()

display(Markdown(f'F1-Score: {metrics.f1_score(true, pred)}'))
display(Markdown(f'Confusion Matrix: {metrics.confusion_matrix(true, pred)}'))
display(Markdown(f'Precision: {metrics.precision_score(true, pred)}'))
display(Markdown(f'Recall: {metrics.recall_score(true, pred)}'))
display(Markdown(f'AUC: {metrics.roc_auc_score(true, pred_prob)}'))

```

#### iii) p-value calculation and refitting

```python
scores, pvalues = chi2(df[feature_cols], df[label])
significant = []
display(Markdown('#### Statistically Significant p-values'))
for i in range(30):
    if pvalues[i] < 0.5 and pvalues[i] > 0:
        significant.append(feature_cols[i])
        display(Markdown(f'**{feature_cols[i]}** p-value: {pvalues[i]}'))

logr_revised = LogisticRegression(random_state=1, solver='liblinear')
logr_revised.fit(train_X[significant], train_y[label])
pred_prob = logr_revised.predict_proba(test_X[significant])[:,1]
fpr, tpr, thresholds = metrics.roc_curve(test_y, pred_prob)
true = test_y

plt.plot([0,1],[0,1], 'k--')
plt.plot(fpr, tpr, label='Logistic Regression')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()

display(Markdown(f'F1-Score: {metrics.f1_score(true, pred)}'))
display(Markdown(f'Confusion Matrix: {metrics.confusion_matrix(true, pred)}'))
display(Markdown(f'Precision: {metrics.precision_score(true, pred)}'))
display(Markdown(f'Recall: {metrics.recall_score(true, pred)}'))
display(Markdown(f'AUC: {metrics.roc_auc_score(true, pred_prob)}'))
```

## Part 2

### a/b/c)

```python
cols = [
    'id',
    'outcome',
    'income'
]
feature_cols = []
label = 'outcome'

for i in range(32):
    cols.append(f'feature_{i}')
    feature_cols.append(f'feature_{i}')
df = pd.read_csv('./wpbc.data', names=cols)
df.drop(columns=['id', 'income'], inplace=True)
df.outcome = df.outcome.map({'R': 1, 'N': 0})
# Change missing feature to median of data set
missing_median =  df[df.feature_31 != '?']['feature_31'].median()
df.feature_31 = df.feature_31.replace({'?': missing_median})

nonrecur = df.loc[df['outcome'] == 0]
recur = df.loc[df['outcome'] == 1]

X = df.copy()
X.drop(columns=['outcome'], inplace=True)
y = df.copy()
y.drop(columns=feature_cols, inplace=True)

train_set = nonrecur[:130].append(recur[:37])
test_set = nonrecur[130:].append(recur[37:])
train_X = train_set.copy().drop(columns=[label])
train_y = train_set.copy().drop(columns=feature_cols)
test_X = test_set.copy().drop(columns=[label])
test_y = test_set.copy().drop(columns=feature_cols)
```

### d)

#### i) Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(train_X, train_y[label])
pred = gnb.predict(test_X)
pred_prob = gnb.predict_proba(test_X)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(test_y, pred_prob)
true = test_y

plt.plot([0,1],[0,1], 'k--')
plt.plot(fpr, tpr, label='Naive Bayes')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()

display(Markdown(f'F1-Score: {metrics.f1_score(true, pred)}'))
display(Markdown(f'Confusion Matrix: {metrics.confusion_matrix(true, pred)}'))
display(Markdown(f'Precision: {metrics.precision_score(true, pred)}'))
display(Markdown(f'Recall: {metrics.recall_score(true, pred)}'))
display(Markdown(f'AUC: {metrics.roc_auc_score(true, pred_prob)}'))
```

#### ii) Naive Bayes with SMOTE

```python
sm = SMOTE(random_state=1, k_neighbors=5, sampling_strategy='all')
X_SMOTE, y_SMOTE = sm.fit_resample(train_X, train_y[label])
X_SMOTE = pd.DataFrame(data=X_SMOTE)
y_SMOTE = pd.DataFrame(data=y_SMOTE)

gnb = GaussianNB()
gnb.fit(X_SMOTE, y_SMOTE[0])
pred = gnb.predict(test_X)
pred_prob = gnb.predict_proba(test_X)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(test_y, pred_prob)
true = test_y

plt.plot([0,1],[0,1], 'k--')
plt.plot(fpr, tpr, label='Naive Bayes')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()

display(Markdown(f'F1-Score: {metrics.f1_score(true, pred)}'))
display(Markdown(f'Confusion Matrix: {metrics.confusion_matrix(true, pred)}'))
display(Markdown(f'Precision: {metrics.precision_score(true, pred)}'))
display(Markdown(f'Recall: {metrics.recall_score(true, pred)}'))
display(Markdown(f'AUC: {metrics.roc_auc_score(true, pred_prob)}'))
display(Markdown(f'**Yes**, SMOTE does help, as seen by the marked increase in F1-Score, Precision, Recall, and the AUC for ROC.'))
```

#### iii) Using multinomial priors


##### Without SMOTE

```python
mnb = MultinomialNB(fit_prior=True)
mnb.fit(train_X, train_y[label])
pred = mnb.predict(test_X)
pred_prob = mnb.predict_proba(test_X)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(test_y, pred_prob)
true = test_y

plt.plot([0,1],[0,1], 'k--')
plt.plot(fpr, tpr, label='Multinomial Naive Bayes')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()

display(Markdown(f'F1-Score: {metrics.f1_score(true, pred)}'))
display(Markdown(f'Confusion Matrix: {metrics.confusion_matrix(true, pred)}'))
display(Markdown(f'Precision: {metrics.precision_score(true, pred)}'))
display(Markdown(f'Recall: {metrics.recall_score(true, pred)}'))
display(Markdown(f'AUC: {metrics.roc_auc_score(true, pred_prob)}'))
```

##### With SMOTE

```python
mnb = MultinomialNB(fit_prior=True)
mnb.fit(X_SMOTE, y_SMOTE[0])
pred = mnb.predict(test_X)
pred_prob = mnb.predict_proba(test_X)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(test_y, pred_prob)
true = test_y

plt.plot([0,1],[0,1], 'k--')
plt.plot(fpr, tpr, label='Multinomial Naive Bayes')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()

display(Markdown(f'F1-Score: {metrics.f1_score(true, pred)}'))
display(Markdown(f'Confusion Matrix: {metrics.confusion_matrix(true, pred)}'))
display(Markdown(f'Precision: {metrics.precision_score(true, pred)}'))
display(Markdown(f'Recall: {metrics.recall_score(true, pred)}'))
display(Markdown(f'AUC: {metrics.roc_auc_score(true, pred_prob)}'))
```
